```python
#Importing required module into our IDE
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Defining subject class with instance variables such as name, hours, completed_topics and total_topics
class Subject:
  def __init__(self,name,hours=0,completed_topics=0,total_topics=0):
    self.name=name
    self.hours=hours if hours is not None else 0
    self.completed_topics=completed_topics if completed_topics is not None else 0
    self.total_topics=total_topics if total_topics is not None else 0
    self.topic_names = []
    self.completed_topic_names = []
```

```python
#Test Case for Creating a subject
math = Subject("Math", hours=2, completed_topics=1, total_topics=5)
print(math.__dict__)
```

```python
#Defining method to add hours for the subject
def add_hours(self,hrs):
  try:
    if hrs <= 0:
      return "âŒ Hours must be positive"
    else:
      self.hours += hrs
      return f"âœ… Added {hrs} hours to {self.name}"
  except Exception as e:
    return f"â— Error: {str(e)}"

#We add the method into a variable add_hours
Subject.add_hours = add_hours
```

```python
#Test case for valid input and invalid input

#Test Case 1
print(math.add_hours(2)) #Valid

#Test Case 2
print(math.add_hours(-2)) #Invalid
```

```python
#Defining a method to set the total amount of topics to be completed
def add_topic_names(self, topics):
  try:
    if not topics or not str(topics).strip():
      return "âŒ Topic list cannot be empty!"
    new_topics = [t.strip() for t in str(topics).split(",") if t.strip()]
    # avoid duplicates when adding
    for t in new_topics:
      if t not in self.topic_names:
        self.topic_names.append(t)
    self.total_topics = len(self.topic_names)
    return f"ðŸ†• Added topics to {self.name}: {', '.join(new_topics)}"
  except Exception as e:
    return f"â— Error: {str(e)}"

#We add the method set_total_topics in a variable set_total_topics
Subject.add_topic_names = add_topic_names
```

```python
#Test case for valid input and invalid input

#Test 1: Valid input
print(math.add_topic_names("Algebra, Geometry, Calculus"))

#Test 2: More Valid input
print(math.add_topic_names("Trigonometry, Algebra"))

#Test 3: Invalid Input
print(math.add_topic_names(""))
```

```python
#Defining method to add the number of topics completed
def complete_topic_names(self, completed):
  try:
    if not completed:
      return "âš ï¸ No topics selected!"
    newly_completed = []
    # completed may come as single string or list â€” normalize to list
    if isinstance(completed, str):
      completed = [completed]
    for t in completed:
      if t in self.topic_names and t not in self.completed_topic_names:
        self.completed_topic_names.append(t)
        newly_completed.append(t)
    self.completed_topics = len(self.completed_topic_names)
    if newly_completed:
      return f"âœ… Marked topics as completed for {self.name}: {', '.join(newly_completed)}"
    else:
      return f"âš ï¸ No new topics were marked (already completed or not found)."
  except Exception as e:
    return f"â— Error: {str(e)}"

#We assign the method complete_topics to a variable complete_topics
Subject.complete_topic_names = complete_topic_names
```

```python
#Test cases for topics completed

print(math.complete_topic_names("Algebra"))                 # Single topic

print(math.complete_topic_names(["Geometry", "Calculus"])) # Multiple

print(math.complete_topic_names("Algebra"))                 # Already completed topic

print(math.complete_topic_names([]))                        # Empty input

print(math.complete_topic_names("Trigonometry"))
```

```python
#Defining a method to get the percentage of progress of a certain subject based off how many topics are completed
def get_progress(self):
  try:
    if self.total_topics==0:
      return "âš ï¸ Total topics not sent"
    else:
      progress = (self.completed_topics / self.total_topics) *100
      return f"ðŸ“Š Progress for {self.name}: {progress:.2f}%"
  except Exception as e:
    return f"â— Error: {str(e)}"

#We assign get_progress method into a variable get_progress
Subject.get_progress = get_progress
```

```python
#Test Case to check the progress if its printing properly or not
print(math.get_progress())
```

```python
#Defining another class for our StudyPlanner
class StudyPlanner:
  def __init__(self):
    self.subjects = {}
```

```python
planner=StudyPlanner()
```

```python
#Defining a method to add the subject based off the name
def add_subject(self,name):
  if not name or name.strip() == "":
    return "âš ï¸ Subject name cannot be empty"
  if name in self.subjects:
    return "âš ï¸ Subject already exists"
  else:
    self.subjects[name] = Subject(name)
    return f"âœ… Subject '{name}' added successfully!"

#We add the method add_subject to a variable add_subject
StudyPlanner.add_subject = add_subject
```

```python
#Test Cases to check whether its Valid, Invalid or Duplicate input

#Test Case 1: To check valid input
print(planner.add_subject("Math")) #Valid

#Test Case 2: To check invalid input
print(planner.add_subject(""))  # Invalid

#Test Case 3: To check duplicate input
print(planner.add_subject("Math")) #Duplicate
```

```python
#Add topic names passthrough
def add_topic_names_planner(self, name, topics):
  if name not in self.subjects:
    return f"âŒ {name} not found"
  return self.subjects[name].add_topic_names(topics)

StudyPlanner.add_topic_names = add_topic_names_planner
```

```python
print(planner.add_topic_names("Math", "Algebra, Geometry"))  # Valid
print(planner.add_topic_names("Math", "Calculus"))           # Valid, adds more
print(planner.add_topic_names("Chemistry", "Organic"))       # Invalid, subject doesn't exist
```

```python
#Add study hours passthrough
def add_study_hours(self, name, hrs):
  if name not in self.subjects:
    return f"âŒ {name} not found"
  return self.subjects[name].add_hours(hrs)

StudyPlanner.add_study_hours = add_study_hours
```

```python
print(planner.add_study_hours("Math", 3))       # Valid
print(planner.add_study_hours("Biology", 1))    # Invalid, subject doesn't exist
print(planner.add_study_hours("Math", -2))      # Invalid input handled by Subject.add_hours
```

```python
#Complete topics by name passthrough (returns message)
def complete_topics_by_name(self, name, topics):
  if name not in self.subjects:
    return f"âŒ {name} not found"
  return self.subjects[name].complete_topic_names(topics)

StudyPlanner.complete_topics_by_name = complete_topics_by_name
```

```python
print(planner.complete_topics_by_name("Math", "Algebra"))           # Single topic
print(planner.complete_topics_by_name("Math", ["Geometry"]))        # List of topics
print(planner.complete_topics_by_name("Math", ["Calculus", "Algebra"])) # Already completed + new
print(planner.complete_topics_by_name("Chemistry", "Organic"))      # Invalid, subject not found
```

```python
#Get progress passthrough
def get_progress_planner(self, name):
  if name not in self.subjects:
    return f"âŒ {name} not found"
  return self.subjects[name].get_progress()

StudyPlanner.get_progress = get_progress_planner
```

```python
print("---Get Progress---")
print(planner.get_progress("Math"))     # Should show progress based on completed topics
print(planner.get_progress("Chemistry"))# Invalid, subject not found
```

```python
#Defining a method to remove the name of the subject
def remove_subject(self,name):
  if name not in self.subjects:
    return f"âŒ Subject '{name}' not found!"
  else:
    del self.subjects[name]
    return f"âœ… Subject {name} removed successfully!"

#We add the method remove_subject to a variable remove_subject
StudyPlanner.remove_subject = remove_subject
```

```python
#Test Cases for valid and invalid input

#Test Case 1: When input is Valid
print(planner.remove_subject("Math")) #Valid

#Test Case 2: When input is Invalid
print(planner.remove_subject("English")) #Invalid
```

```python
#Defining a method to generate a report by summarizing all the information that we got based off the previous methods
def generate_report(self):
    if not self.subjects:
        return "âš ï¸ No subjects added yet!"

    report_lines = []

    for sub in self.subjects.values():
        # Ensure total_topics is not zero to avoid division by zero
        progress = (sub.completed_topics / sub.total_topics * 100) if sub.total_topics else 0
        report_lines.append(
            f"ðŸ“˜ {sub.name}\n"
            f"â±ï¸ Hours Studied: {sub.hours}\n"
            f"âœ… Topics Completed: {sub.completed_topics}/{sub.total_topics}\n"
            f"ðŸ“Š Progress: {progress:.2f}%\n"
            "------------------"
        )

    # Join all lines into a single string
    return "\n".join(report_lines)

#We assign the method generate_report to a variable generate_report
StudyPlanner.generate_report = generate_report
```

```python
#Test Case to check whether the report is being generated or not
print(planner.generate_report())
```

```python
#Convert subjects to Pandas DataFrame
def to_dataframe(self):
    data = []
    for sub in self.subjects.values():
        progress = (sub.completed_topics / sub.total_topics * 100) if sub.total_topics else 0
        data.append({
            "Subject": sub.name,
            "Hours": sub.hours,
            "Completed": sub.completed_topics,
            "Total": sub.total_topics,
            "Progress %": round(progress, 2),
            "Topics": ", ".join(sub.topic_names),
            "Completed Topics": ", ".join(sub.completed_topic_names)
        })
    return pd.DataFrame(data)

StudyPlanner.to_dataframe = to_dataframe
```

```python
# ---- NEW: Matplotlib helpers that return Figures ----
def plot_progress_bar_fig(self):
    df = self.to_dataframe()
    if df.empty:
        fig, ax = plt.subplots(figsize=(6,3))
        ax.text(0.5, 0.5, "No data available", ha="center", va="center")
        ax.axis("off")
        return fig
    fig, ax = plt.subplots(figsize=(7,4))
    ax.bar(df['Subject'], df['Progress %'])
    ax.set_ylabel('Progress (%)')
    ax.set_title('Study Progress by Subject')
    ax.set_ylim(0,100)
    for i, v in enumerate(df['Progress %']):
        ax.text(i, v + 1, f"{v:.1f}%", ha='center')
    fig.tight_layout()
    return fig

StudyPlanner.plot_progress_bar_fig = plot_progress_bar_fig
```

```python
def plot_hours_fig(self):
    df = self.to_dataframe()
    if df.empty:
        fig, ax = plt.subplots(figsize=(6,3))
        ax.text(0.5, 0.5, "No data available", ha="center", va="center")
        ax.axis("off")
        return fig
    fig, ax = plt.subplots(figsize=(7,4))
    ax.bar(df['Subject'], df['Hours'])
    ax.set_ylabel('Hours Studied')
    ax.set_title('Hours Studied by Subject')
    for i, v in enumerate(df['Hours']):
        ax.text(i, v + 0.1, f"{v}", ha='center')
    fig.tight_layout()
    return fig

StudyPlanner.plot_hours_fig = plot_hours_fig
```

```python
def plot_topics_completed_fig(self):
    df = self.to_dataframe()
    if df.empty:
        fig, ax = plt.subplots(figsize=(6,3))
        ax.text(0.5, 0.5, "No data available", ha="center", va="center")
        ax.axis("off")
        return fig
    fig, ax = plt.subplots(figsize=(7,4))
    ax.bar(df['Subject'], df['Completed'])
    ax.set_ylabel('Completed Topics')
    ax.set_title('Topics Completed by Subject')
    for i, v in enumerate(df['Completed']):
        ax.text(i, v + 0.1, f"{v}", ha='center')
    fig.tight_layout()
    return fig

StudyPlanner.plot_topics_completed_fig = plot_topics_completed_fig
```

```python
#Create a bar chart showing progress for each subject
def plot_progress(self):
    df = self.to_dataframe()
    if df.empty:
        print("âš ï¸ No data available.")
        return

    plt.figure(figsize=(7,5))
    plt.bar(df["Subject"], df["Progress %"])
    plt.ylabel("Progress (%)")
    plt.title("Study Progress by Subject")
    plt.xticks(rotation=45)
    plt.show()

#We assign the method plot_progress to a variable plot_progress
StudyPlanner.plot_progress = plot_progress
```

```python
#Gradio UI by taking theme as soft (cuz the color was cool and aesthetic).

with gr.Blocks(theme=gr.themes.Soft()) as app: # Creating a space called app for our UI components (eg: tabs. buttons, textboxes)
    gr.Markdown("# ðŸŽ“ Student Study Planner & Progress Tracker") # Markdown is used for the title

    # Add Subject
    with gr.Tab("ðŸ“˜ Add Subject"): # Creates a tab called add subject
        s_name = gr.Textbox(label="Subject Name")  # Creates a textbox with the title as Subject Name stored in a variable called s_name
        s_btn = gr.Button("âž• Add Subject") # Creates a button called Add Subject where interact = True by default
        s_out = gr.Textbox(label="Output") # Creates a textbox with the title as output to show that the output will be displayed here
        s_btn.click(lambda name: planner.add_subject(name),
                    [s_name], s_out) # lambda is a function that is used when we want to process inputs and return outputs which in this case is name.
                                     # Only happens when the button is clicked thats why the func s_btn.click is there :P

    # Remove Subject
    with gr.Tab("ðŸ—‘ï¸ Remove Subject"): # Creates a tab called Remove Subject
        r_name = gr.Textbox(label="Subject Name to Remove") # Creates a Textbox with title "Subject Name to Remove"
        r_btn = gr.Button("Remove Subject") # Creates a button called Remove Subject
        r_out = gr.Textbox(label="Output") # Creates a textbox with title Output
        r_btn.click(lambda name: planner.remove_subject(name), r_name, r_out) # Uses lambda to take the input name and delete it when the button is clicked

    # Add Study Hours
    with gr.Tab("â±ï¸ Add Study Hours"): # Creates a tab called Add Study Hours
        h_sub = gr.Textbox(label="Subject") # Creates a textbox with title Subject
        h_hours = gr.Number(label="Hours Studied") # Creates a box where you can input numbers with the title Hours Studied
        h_btn = gr.Button("Submit Hours") # Creates a button called Submit Hours
        h_out = gr.Textbox(label="Output") # Creates a textbox with the title Output
        h_btn.click(lambda sub,hrs: planner.add_study_hours(sub, hrs) if sub in planner.subjects else f"âŒ {sub} not found",
                    [h_sub, h_hours], h_out) # When the button is clicked takes input of sub and hrs and adds the hours for the subject

    with gr.Tab("ðŸ“š Add Topics"): # Creates a tab called Add Topics
        t_sub = gr.Textbox(label="Subject") # Textbox to enter subject name
        t_topics = gr.Textbox(label="Enter Topics (comma separated)") # Textbox to enter topics separated by commas
        t_btn = gr.Button("Add Topics") # Button to add entered topics to the subject
        t_out = gr.Textbox(label="Output") # Textbox to display feedback/output

        # Safe wrapper: ensure subject exists before calling add_topic_names
        def add_topics_ui(sub, topics):
            if not sub:
                return "âŒ Subject required" # Error if subject is empty
            if sub not in planner.subjects:
                return f"âŒ {sub} not found" # Error if subject does not exist
            return planner.add_topic_names(sub, topics) # Add topics and return result

        # Bind button click to add_topics_ui function
        t_btn.click(add_topics_ui, [t_sub, t_topics], t_out)

    # Complete Topics
    with gr.Tab("âœ… Complete Topics"): # Creates a tab called Complete Topics
        c_sub = gr.Textbox(label="Subject") # Textbox to enter subject name
        c_refresh = gr.Button("ðŸ”„ Load Topics") # Button to refresh/load the topics for the selected subject
        c_topics = gr.Dropdown(label="Select Topics", choices=[], multiselect=True) # Dropdown to select multiple topics, initially empty
        c_btn = gr.Button("Mark Complete") # Button to mark selected topics as completed
        c_out = gr.Textbox(label="Output") # Textbox to display feedback/output

        # Function to update dropdown choices dynamically
        # Return gr.update to update the dropdown with topics that are not yet completed
        def load_topics(sub):
            try:
                if sub and sub in planner.subjects:
                    # Show only not-yet-completed topics in the dropdown
                    choices = [t for t in planner.subjects[sub].topic_names if t not in planner.subjects[sub].completed_topic_names]
                    return gr.update(choices=choices, value=[])
                # If subject is empty or not found, return empty dropdown
                return gr.update(choices=[], value=[])
            except Exception as e:
                # On error, return empty dropdown
                return gr.update(choices=[], value=[])

        # Function to mark selected topics as completed and refresh dropdown choices
        def mark_complete_and_refresh(sub, topics):
            try:
                if not sub:
                    return ("âŒ Subject required", gr.update(choices=[], value=[])) # Error if subject is empty
                if sub not in planner.subjects:
                    return (f"âŒ {sub} not found", gr.update(choices=[], value=[])) # Error if subject not found

                # Normalize topics parameter to a list
                if topics is None:
                    return ("âš ï¸ No topics selected", gr.update(
                        choices=[t for t in planner.subjects[sub].topic_names if t not in planner.subjects[sub].completed_topic_names],
                        value=[]
                    ))
                if isinstance(topics, str):
                    topics = [topics]
                topics = list(topics)

                # Call method to mark topics as completed
                msg = planner.complete_topics_by_name(sub, topics)

                # After marking, remove completed topics from dropdown choices
                choices = [t for t in planner.subjects[sub].topic_names if t not in planner.subjects[sub].completed_topic_names]
                return (msg, gr.update(choices=choices, value=[]))
            except Exception as e:
                # On error, return empty dropdown and error message
                return (f"â— Error: {e}", gr.update(choices=[], value=[]))

        # Bind refresh button to load topics function
        c_refresh.click(load_topics, inputs=c_sub, outputs=c_topics)
        # Bind mark complete button to mark_complete_and_refresh function
        c_btn.click(mark_complete_and_refresh, inputs=[c_sub, c_topics], outputs=[c_out, c_topics])

    # View Progress
    with gr.Tab("ðŸ“Š View Progress"): # Creates a tab called View Progress
        p_sub = gr.Textbox(label="Subject") # Creates a textbox called Subject
        p_btn = gr.Button("Show Progress") # Creates a button with the label Show Progress
        p_out = gr.Textbox(label="Progress Result") # Creates a textbox with the title Progress Result (basically the output :P)
        p_btn.click(lambda sub: planner.get_progress(sub) if sub in planner.subjects else f"âŒ {sub} not found",
                    p_sub, p_out) # When the button is pressed it takes the subject and runs the method get_progress and displays the result

    # Charts & Data
    with gr.Tab("ðŸ“ˆ Charts & Data"):
        d_btn = gr.Button("Generate DataFrame & Charts")
        d_table = gr.Dataframe(headers=["Subject","Hours","Completed","Total","Progress %","Topics","Completed Topics"], datatype=["str","number","number","number","number","str","str"], interactive=False)
        csv_file = gr.File(label="Download CSV")
        img_progress = gr.Plot(label="Progress Chart")
        img_hours = gr.Plot(label="Hours Chart")
        img_topics = gr.Plot(label="Topics Completed Chart")

        def make_df_and_plots():
            df = planner.to_dataframe()
            # ensure df is a DataFrame even if empty
            if df is None or df.empty:
                empty_df = pd.DataFrame(columns=["Subject","Hours","Completed","Total","Progress %","Topics","Completed Topics"])
                # create blank figs
                fig_blank = plt.figure(figsize=(6,3))
                plt.text(0.5, 0.5, "No data available", ha="center", va="center")
                plt.axis('off')
                # Save an empty CSV
                csv_name = "study_planner_report.csv"
                empty_df.to_csv(csv_name, index=False)
                return (empty_df, fig_blank, fig_blank, fig_blank, csv_name)
            # Generate plots
            fig_p = planner.plot_progress_bar_fig()
            fig_h = planner.plot_hours_fig()
            fig_t = planner.plot_topics_completed_fig()
            # Save CSV to temporary file for download
            csv_name = "study_planner_report.csv"
            df.to_csv(csv_name, index=False)
            return (df, fig_p, fig_h, fig_t, csv_name)

        d_btn.click(make_df_and_plots, None, [d_table, img_progress, img_hours, img_topics, csv_file])

    # Full Report
    with gr.Tab("ðŸ“‘ Full Report"): # Creates a tab called Full Report
        r_btn = gr.Button("Generate Report") # Creates a button with label Generate Report
        r_out = gr.Textbox(label="Report", lines=10) # Creates a textbox with the title Report and number of lines that will be visible are 10
        r_btn.click(lambda: planner.generate_report(), None, r_out) # When button is clicked runs the generate report function and displays the output

    # Utility & Quick Actions
    with gr.Tab("âš™ï¸ Utilities"):
        sample_btn = gr.Button("Create Sample Data")
        sample_out = gr.Textbox(label="Output")
        def create_sample():
            # quick sample data creation for testing the UI
            planner.subjects = {}
            planner.add_subject("Math")
            planner.add_topic_names("Math", "Algebra, Geometry, Calculus")
            planner.add_study_hours("Math", 5)
            planner.complete_topics_by_name("Math", ["Algebra", "Calculus"])
            planner.add_subject("Physics")
            planner.add_topic_names("Physics", "Kinematics, Dynamics")
            planner.add_study_hours("Physics", 2)
            planner.complete_topics_by_name("Physics", ["Kinematics"])
            return "âœ… Sample data created."
        sample_btn.click(create_sample, None, sample_out)

        clear_btn = gr.Button("Reset Planner (Deletes all subjects)")
        clear_out = gr.Textbox(label="Output")
        def reset_planner():
            planner.subjects = {}
            return "âœ… Planner reset. All subjects cleared."
        clear_btn.click(reset_planner, None, clear_out)

app.launch()
```
